```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        print(height)
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, -10),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
def autolabelError(rects, errors):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for i in range(len(rects)):
        rect = rects[i]
#       height = rect.get_height()
        height = errors[i]
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 0),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
def plotReads():
    cost = [3.253, 5.412]
    nn = [5.924, 11.020]
    error = [(0, 0), [5.924 - cost[0], 11.020 - cost[1]]]
    store = ('DynanoDB', 'AFTSI')
    ind = np.arange(2)
    colors = ['teal','seagreen']
    barList = plt.bar(ind, cost, yerr=error, color=colors)
    plt.title('Read Latency')
    plt.xticks(ind, store)
    autolabel(barList)
    autolabelError(barList, nn)
    plt.ylabel('Latency (ms)')
    #plt.show()
    plt.savefig('Read421.png')
    
```

```python
plotReads()
```

```python
def plotCom():
    cost = [3.202, 14.387]
    nn = [628.829, 97.574]
    error = [(0, 0), [628.829 - cost[0], 97.574 - cost[1]]]
    store = ('DynanoDB Write', 'AFTSI Commit')
    ind = np.arange(2)
    colors = ['teal','seagreen']
    barList = plt.bar(ind, cost, yerr=error, color=colors)
    plt.title('Transaction Latency')
    plt.xticks(ind, store)
    autolabel(barList)
    autolabelError(barList, nn)
    plt.ylabel('Latency (ms)')
    plt.yscale('log')
    #plt.show()
    plt.savefig('Com421.png')
```

```python
plotCom()
```

```python
def stacked():
    N = 1
    a = (3)
    b = (4)
    c = (4)
    d = (3)
    ind = np.arange(N)    # the x locations for the groups
    #width = 0.35       # the width of the bars: can also be len(x) sequence

    p1 = plt.bar(ind, a)
    p2 = plt.bar(ind, b, bottom=a)
    p3 = plt.bar(ind, c, bottom=b)
    p4 = plt.bar(ind, d, bottom=c)
    

    plt.ylabel('Latency (ms)')
    plt.title('Commit Breakdown')
    plt.xticks(ind, ('AFTSI'))
    plt.legend((p1[0], p2[0], p3[0], p4[0]), ('Phase 1: Validation', 'Write to Storage', 'Phase 2', 'GRPC Overhead'))

    #plt.show()
    plt.savefig('421commitbreakdown.png')
```
